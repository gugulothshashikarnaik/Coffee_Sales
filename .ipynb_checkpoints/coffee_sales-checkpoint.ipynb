```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#load dataset
file_path = "/Users/shashiroy/Desktop/coffee_sales/index.csv"
coffee_data = pd.read_csv('index.csv')
```

```python
#EDA
coffee_data.head()
coffee_data.info()
missing_values = coffee_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
duplicates_count = coffee_data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_count}")
print(coffee_data.describe().T)
print(coffee_data.loc[:,['cash_type','card','coffee_name']].describe().T)
cash_type_null_card = coffee_data[coffee_data['card'].isnull()]['cash_type'].value_counts()
print("Cash type counts where card is null:\n", cash_type_null_card)
coffee_data['cash_type'].hist()
coffee_data['cash_type'].value_counts().plot(kind='bar')


# Plotting histogram for cash type distribution
plt.figure(figsize=(10, 5))
coffee_data['cash_type'].hist( bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Cash Type')
plt.xlabel('Cash Type')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Plotting bar chart for cash type value counts
plt.figure(figsize=(10, 5))
coffee_data['cash_type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Cash Type Counts')
plt.xlabel('Cash Type')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# DataFrame showing the normalized counts of coffee names
coffee_name_distribution = pd.DataFrame(coffee_data['coffee_name'].value_counts(normalize = True).sort_values(ascending=False).round(4)*100)

# Display the DataFrame
print(coffee_name_distribution)
```

```python
#convert date and datetime to datetme format
coffee_data['date']=pd.to_datetime(coffee_data['date'], errors='coerce')
coffee_data['datetime']=pd.to_datetime(coffee_data['datetime'], errors='coerce')
#Create column of month, weekdays, and Hours
coffee_data['month']=coffee_data['date'].dt.strftime('%Y-%m')
coffee_data['day']=coffee_data['date'].dt.strftime('%A')
coffee_data['hour']=coffee_data['datetime'].dt.hour
print(coffee_data.head())
```

```python
[coffee_data['date'].min(),coffee_data['date'].max()]
```

```python
revenue_data = coffee_data.groupby(['coffee_name']).sum(['money']).reset_index().sort_values(by='money' ,ascending=False)
```

```python
plt.figure(figsize=(10,4))
ax = sns.barplot(data=revenue_data,x='money' ,y='coffee_name' ,color='steelblue')
ax.bar_label(ax.containers[0], fontsize=6)
plt.xlabel('Revenue')
plt.ylabel('Coffee Name')
plt.title('Revenue by Coffee Type')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = coffee_data.groupby(['coffee_name' , 'month']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='month',columns='coffee_name' ,values='count').reset_index()
monthly_sales
```

```python
monthly_sales.describe().T.loc[:,['min','max']]
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales.set_index('month'))
plt.legend(loc='upper left')
plt.xticks(range(len(monthly_sales['month'])),monthly_sales['month'],rotation=45, ha='right')
plt.xlabel('Month')
plt.ylabel('Sales Count')
plt.title('Monthly Sales of Coffee Types')
plt.tight_layout()
plt.show()
```

```python
weekday_sales = coffee_data.groupby(['day']).count()['date'].reset_index().rename(columns={'date':'count'})

#Show the resulting DataFrame
weekday_sales

```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=weekday_sales,x='day', y='count' ,color='steelblue')
plt.xticks(ticks=range(len(weekday_sales['day'])), labels=['Sun','Mon','Tue','Wed','Thur','Fri','Sat'],size='small')
plt.xlabel('Day of the Week')
plt.ylabel('Sales Count')
plt.title('Coffee Sales Count by Day of the Week')
plt.tight_layout()
plt.show()
```

```python
daily_sales = (coffee_data.groupby(['coffee_name', 'date']).count()['datetime'].reset_index().rename(columns={'datetime': 'count'}))
daily_sales = daily_sales.pivot(index='date', columns='coffee_name', values='count').reset_index().fillna(0)
daily_sales
```

```python
min_max_sales = daily_sales.iloc[:, 1:].describe().T.loc[:, ['min', 'max']]
min_max_sales
```

```python
# Convert 'date' and 'datetime' columns to datetime objects
coffee_data['date'] = pd.to_datetime(coffee_data['date'])
coffee_data['datetime'] = pd.to_datetime(coffee_data['datetime'])

# Extract hour from 'datetime' column
coffee_data['hour'] = coffee_data['datetime'].dt.hour

# Group by hour and count occurrences
hourly_sales = coffee_data.groupby('hour').count()['datetime'].reset_index().rename(columns={'datetime': 'count'})
print(hourly_sales)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=hourly_sales,x='hour',y='count',color='steelblue')
plt.xlabel('Hour of the Day')
plt.ylabel('Sales Count')
plt.title('Hourly Coffee Sales')
plt.xticks(range(24))  # Ensure all hours from 0 to 23 are shown
plt.tight_layout()  # Adjusts the layout for better fit
plt.show()
```

```python
hourly_sales_by_coffee = (coffee_data.groupby(['hour','coffee_name']).count()['date'].reset_index().rename(columns={'date':'count'}))
hourly_sales_by_coffee = (hourly_sales_by_coffee.pivot(index='hour', columns='coffee_name', values='count').fillna(0).reset_index())
hourly_sales_by_coffee
```

```python
# Create a figure with subplots (2 rows, 4 columns)
fig, axs = plt.subplots(2, 4, figsize=(20, 10))

# Flatten the array of subplots for easy iteration
axs = axs.flatten()

# Loop through each coffee type, skipping the 'hour' column
for i, column in enumerate(hourly_sales_by_coffee.columns[1:]):  # Skip the first column ('hour')
    axs[i].bar(hourly_sales_by_coffee['hour'], hourly_sales_by_coffee[column], color='steelblue')
    axs[i].set_title(f'Sales of {column} by Hour')
    axs[i].set_xlabel('Hour')
    axs[i].set_ylabel('Sales Count')  # Add label for the y-axis
    axs[i].set_xticks(range(24))  # Ensure all hours are displayed on the x-axis

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
